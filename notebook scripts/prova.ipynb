```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Main Circle
theta_main = np.linspace(0, 2*np.pi, 400)

# 2. Define the Arc Range
# Let's say the indent happens between 0 and 90 degrees (0 to pi/2)
arc_start = 0
arc_end = np.pi/2

# 3. Create the path by modifying the radius based on the angle
# r = 1.0 - 0.7 * np.exp(-((theta_main - np.pi/4)**2) / 0.1)

r = 1
x = r * np.cos(theta_main)
y = r * np.sin(theta_main)

angles = np.arctan2(y, x)
sort_idx = np.argsort(angles)
x_sorted = x[sort_idx]
y_sorted = y[sort_idx]

area = 0
for i in range(0,399):
    area = area + (y_sorted[i]+y_sorted[i+1])*(x_sorted[i]-x_sorted[i+1])

print(np.pi*r**2)
print(1/2*area)

# 4. Plot
plt.figure(figsize=(6,6))
plt.plot(x, y)
plt.axis('equal')
plt.show()
```
